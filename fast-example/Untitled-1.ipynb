```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a new figure and an axes object for the 3D plot.
# 'projection="3d"' is the key to creating a 3D plot.


df = pd.read_csv("ac_1sec.csv")
df
# --- Trajectory 1: A spiral path ---
# Generate data for the first trajectory.
# We'll create a spiral shape by using sine and cosine functions for x and y,
# and a linear progression for the z-axis.
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


# Plot the first trajectory.
ax.plot(df["f1_lat"], df["f1_lon"], df["f1_alt_ft"], label='Trajectory 1 (Spiral)', color='blue', linestyle='--')

# --- Trajectory 2: A different spiral path ---
# Generate data for the second trajectory.
# This one will have a different radius and a more jagged progression on the z-axis
# to make it distinct.
# t2 = np.linspace(-10, 10, 500)
# x2 = 2 * np.cos(t2)
# y2 = 2 * np.sin(t2)
# z2 = np.sin(2 * t2) + t2

# Plot the second trajectory.
ax.plot(df["f2_lat"], df["f2_lon"], df["f2_alt_ft"], label='Trajectory 2 (Wavy Spiral)', color='red', linewidth=2, linestyle="--")

# Set the axis labels to make the plot clear.
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add a title to the plot.
ax.set_title('Two 3D Trajectories on a Single Plot')

# Add a legend to distinguish between the two trajectories.
ax.legend()

# Display the plot.
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# --- Trajectory 1: A spiral path with cubes ---
# Plot the line trajectory.
ax.plot(df["f1_lat"][::3], df["f1_lon"][::3], df["f1_alt_ft"][::3], label='Trajectory 1 (Line)', color='blue', linestyle='-')
# Plot the cubes (scatter points).
ax.scatter(df["f1_lat"][::3], df["f1_lon"][::3], df["f1_alt_ft"][::3], label='Trajectory 1 (Cubes)', color='cyan', marker='s', s=100)

# --- Trajectory 2: A different spiral path with cubes ---
# Plot the line trajectory.
ax.plot(df["f2_lat"][::3], df["f2_lon"][::3], df["f2_alt_ft"][::3], label='Trajectory 2 (Line)', color='red', linestyle='-')
# Plot the cubes (scatter points).
ax.scatter(df["f2_lat"][::3], df["f2_lon"][::3], df["f2_alt_ft"][::3], label='Trajectory 2 (Cubes)', color='magenta', marker='s', s=100)

# Set the axis labels to make the plot clear.
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add a title to the plot.
ax.set_title('Two 3D Trajectories with Cubes')

# Add a legend to distinguish between the two trajectories.
ax.legend()

# Display the plot.
plt.show()

```
