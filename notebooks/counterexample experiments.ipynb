```python
# Generate grid of points
x_vals = np.arange(-2, 2, 0.05)
y_vals = np.arange(-2, 8, 0.05)
X, Y = np.meshgrid(x_vals, y_vals)

# Classify points
above_t1 = Y > t1(X)
above_t2 = Y > t2(X)

# Combine classifications
conditions = above_t1.astype(int) * 2 + above_t2.astype(int)  # 0 to 3
labels = {
    0: "below both",
    1: "below t1, above t2",
    2: "above t1, below t2",
    3: "above both"
}

df = pd.DataFrame({
    "x": X.ravel(),
    "y": Y.ravel(),
    "zone": [labels[z] for z in conditions.ravel()]
})

# Plot all combinations
fig = px.scatter(df, x="x", y="y", color="zone", opacity=0.4,
                 category_orders={"zone": list(labels.values())},
                 labels={"zone": "Region"})

# Add both curves
x_curve = x_vals
fig.add_trace(go.Scatter(x=x_curve, y=t1(x_curve), mode='lines',
                         name='y = (x + 2)² + 1', line=dict(color="black")))
fig.add_trace(go.Scatter(x=x_curve, y=t2(x_curve), mode='lines',
                         name='y = (x - 2)² - 1', line=dict(color="black", dash="dash")))

fig.update_layout(title="Regions Defined by Two Curves",
                  legend_title="Region")

fig.add_shape(
    type="rect",
    x0=-2, x1=2,  # width = 4
    y0=-1, y1=1,  # height = 2
    line=dict(color="red", width=2, dash="dot"),
    fillcolor="rgba(0,0,0,0)",  # transparent fill
    layer="above",
)

fig.show()
```
